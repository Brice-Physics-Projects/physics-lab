## Problem 1.1 - Volume of a Cylinder
pg. 17

### Problem Statement
- Nuclear waste material in a physics lab is stored in a cylinder of height 4-13/16 feet and circumference 8-3/16 feet. What is the volume of the cylinder in metric units?

### Given
- h = 4.83125 feet
- c = 8.1875 feet

### Research
- Convert feet to meters
- Use the formula for the volume of a cylinder: V = πr^2h
- To find the radius, use the formula for the circumference of a circle: c = 2πr
- To convert feet to meters, use the conversion factor 1 foot = 0.3048 meters
- Plug the values into the formulas to find the volume in cubic meters

### Calculations
Radius:

$$
r = \frac{C}{2\pi}
$$

Convert to meters:

$$
r_m = r \times 0.3048
$$

$$
h_m = h \times 0.3048
$$

Volume:

$$
V=\pi r^2 h
$$

```python
# Calculation
circumference = 8.1875
h = 4.83125
r = circumference / (np.pi * 2)
r_meters = r * 0.3048
h_meters = h * 0.3048
v = np.pi * r_meters**2*h_meters

print(f'The volume of the cylinder is {v:.2f} cubic meters.')
```

# Problem 1.2 - Volume of a Barrel of Oil
pg. 19

## Problem Statement
The volume of a barrel of oil is 159 liters. We need to design a cylindrical container that will hold this volume of oil.  The container needs to have a height of 1.00 meters to fit in a transport container. What is the required radius of the container?  What is the circumference of the cylindrical container?

## Given
V = 159 liters<br>
h = 1.00 meters

## Research
circumference:
$$
c = 2\pi r
$$

Volume:
$$
V=\pi r^2 h
$$

## Calculations

```python
# Calculations
V = 159
h = 1.00
r = np.sqrt(V/(np.pi*h))
c = 2 * np.pi * r

print(f'The radius of the container is {r:.2f} meters.')
print(f'The circumference of the container is {c:.2f} meters.')
```
